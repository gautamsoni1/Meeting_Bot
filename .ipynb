```python
import json
import os
from groq import Groq

def summarize_json(file_path: str) -> str:
    """
    Load JSON from local file and return summary using Groq Gemma model
    """

    # Initialize Groq client
    client = Groq(api_key=os.getenv("GROQ_API_KEY"))

    # Load JSON file
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except Exception as e:
        return f"Error loading JSON: {str(e)}"

    # Convert JSON to readable text
    text_data = json.dumps(data, indent=2)

    # Optional: truncate if too large (Gemma has ~8k context)
    text_data = text_data[:6000]

    # Call Groq API
    try:
        response = client.chat.completions.create(
            model="llama-3.1-8b-instant",  # Gemma model
            messages=[
                {
                    "role": "system",
                    "content": "You are an AI that summarizes JSON data into clear and concise text."
                },
                {
                    "role": "user",
                    "content": f"Summarize the following JSON data:\n{text_data}"
                }
            ],
            temperature=0.3,
            max_completion_tokens=500
        )

        return response.choices[0].message.content

    except Exception as e:
        return f"Error during summarization: {str(e)}"
```

```python
summary = summarize_json("data.json")
print(summary)
```
